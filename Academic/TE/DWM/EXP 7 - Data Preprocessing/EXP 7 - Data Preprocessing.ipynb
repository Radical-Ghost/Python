```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =  pd.read_csv('zomato.csv', encoding='latin-1') # Read the data
df.sample(5) # Display 5 random rows
```

```python
df.info() # Display the information about the dataset
```

```python
df.describe() # Display the statistical summary of the dataset
```

```python
df.shape # Display the shape of the dataset
```

```python
df.isnull().sum() # Display the number of missing values in the dataset
```

```python
sns.heatmap(df.isnull(), cbar=False) # Display the missing values in the dataset
#due to it being too large, we can't see the missing values in the dataset
```

```python
#missing values in the dataset
missing = [i for i in df.columns if df[i].isnull().sum() > 0]
print(missing)
```

```python
print(df.duplicated()) # Display the duplicated values in the dataset
print("\nDropped Duplicates: ", df.drop_duplicates(inplace=True)) # Drop the duplicated values in the dataset
```

```python
df_numbers_only = df.select_dtypes(include=[np.number]) # Select only the numerical columns
df_numbers_only.sample(5)
```

```python
df_numbers_only.corr() # Display the correlation between the numerical columns
```

```python
sns.heatmap(df_numbers_only.corr(), annot=True) # Display the correlation between the numerical columns
```

```python
Country = pd.read_excel("Country-Code.xlsx")  # Read the country code
Country.sample(5)
```

```python
Country.info()
```

```python
comb_pd = pd.merge(df, Country, left_on='Country Code', right_on='Country Code') # Merge the two datasets
comb_pd.sample(5)
```

```python
print(comb_pd.Country.value_counts()) # Display the number of restaurants in each country
country_names = comb_pd.Country.value_counts().index 
country_values = comb_pd.Country.value_counts().values 
```

```python
plt.pie(country_values, labels=country_names, autopct='%1.1f%%') # Display the number of restaurants in each country
plt.show()
```

```python
plt.pie(country_values[:3], labels=country_names[:3]) # Display the number of restaurants in the top 5 countries
plt.show()
```
